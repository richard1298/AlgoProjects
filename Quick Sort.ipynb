```python
#this is my implementation of quick sort 
def quick_sort_firstelem(my_list,left_bd, right_bd):
    dim = right_bd-left_bd+1
    #dim = len(my_list)
    count = dim-1
    if dim>1:
#        print("handle of the list = ", my_list[left_bd:right_bd+1])
        i = left_bd + 1
        pivot = my_list[left_bd]
#        print("pivot = ",pivot)
        for j in range(left_bd+1,right_bd+1):
            if (my_list[j]<pivot):
#                print("*****one swap happens:")
                my_list[j],my_list[i] = my_list[i], my_list[j]
                i = i + 1
#                print("after this swap: ",my_list[left_bd:right_bd+1])
        my_list[left_bd], my_list[i-1] = my_list[i-1], my_list[left_bd]
#        print("swapping the pivot into its own position:", my_list[left_bd:right_bd+1])
        #i is the index of the first element that is bigger than the pivot
        #i-1 is the index of the pivot after the swap
        if i-1 == left_bd:
#            print("pivot is the smallest in the list, recursion sublist = ",my_list[left_bd+1:right_bd+1])
            right_count = quick_sort_firstelem(my_list,left_bd+1,right_bd)
            count+= right_count
        elif i-1 == right_bd:
#            print("pivot is the biggest in the list, recursion sublist = ",my_list[left_bd:right_bd])
            left_count = quick_sort_firstelem(my_list,left_bd,right_bd-1)
            count+= left_count
        else:
            part_list_left_start = left_bd
            part_list_left_end = i-2
            part_list_right_start = i
            part_list_right_end = right_bd
#            print("left half of partition = ", my_list[part_list_left_start:part_list_left_end+1])
#            print("left half of partition = ", my_list[part_list_right_start:part_list_right_end+1])
            left_count = quick_sort_firstelem(my_list,part_list_left_start,part_list_left_end)
            right_count = quick_sort_firstelem(my_list,part_list_right_start,part_list_right_end)
            count+= (left_count + right_count)
    return count
    
```

```python
#this code is from google
def partition(array, begin, end):
    pivot = begin
    for i in range(begin+1, end+1):
        if array[i] <= array[begin]:
            pivot += 1
            array[i], array[pivot] = array[pivot], array[i]
    array[pivot], array[begin] = array[begin], array[pivot]
    return pivot



def quicksort(array, begin=0, end=None):
    if end is None:
        end = len(array) - 1
    def _quicksort(array, begin, end):
        if begin >= end:
            return
        pivot = partition(array, begin, end)
        _quicksort(array, begin, pivot-1)
        _quicksort(array, pivot+1, end)
    return _quicksort(array, begin, end)
```

```python
def quick_sort_lastelem(my_list,left_bd, right_bd):
    dim = right_bd-left_bd+1
    #dim = len(my_list)
    count = dim-1
#    print(my_list[left_bd:right_bd+1])
#    print(count)
    if dim>1:
        i = left_bd + 1
        my_list[left_bd], my_list[right_bd] = my_list[right_bd],my_list[left_bd]
        pivot = my_list[left_bd]
        for j in range(left_bd+1,right_bd+1):
            if (my_list[j]<pivot):
                my_list[j],my_list[i] = my_list[i], my_list[j]
                i = i + 1
        my_list[left_bd], my_list[i-1] = my_list[i-1], my_list[left_bd]
        #i is the index of the first element that is bigger than the pivot
        #i-1 is the index of the pivot after the swap
        if i-1 == left_bd:
            right_count = quick_sort_lastelem(my_list,left_bd+1,right_bd)
            count+= right_count
        elif i-1 == right_bd:
            left_count = quick_sort_lastelem(my_list,left_bd,right_bd-1)
            count+= left_count
        else:
            part_list_left_start = left_bd
            part_list_left_end = i-2
            part_list_right_start = i
            part_list_right_end = right_bd
            left_count = quick_sort_lastelem(my_list,part_list_left_start,part_list_left_end)
            right_count = quick_sort_lastelem(my_list,part_list_right_start,part_list_right_end)
            count+= (left_count + right_count)
    return count
    
```

```python
import numpy as np
def quick_sort_medelem(my_list,left_bd, right_bd):
    dim = right_bd-left_bd+1
    #dim = len(my_list)
    count = dim-1
#    print(my_list[left_bd:right_bd+1])
#    print(count)
    if dim>1:
        i = left_bd + 1
        #this block of code finds the medium and swap with the first
        firstelem = my_list[left_bd]
        lastelem = my_list[right_bd]
        medidx = left_bd
        if dim%2==1:
            medidx = left_bd + (right_bd-left_bd+1)//2
        else:
            medidx = left_bd + (right_bd-left_bd)//2
        medelem = my_list[medidx]
        pivot_idx = left_bd
        if firstelem == np.median([firstelem,lastelem,medelem]):
            pivot_idx = left_bd
        elif lastelem == np.median([firstelem,lastelem,medelem]):
            pivot_idx = right_bd
        elif medelem == np.median([firstelem,lastelem,medelem]):
            pivot_idx = medidx
        my_list[left_bd], my_list[pivot_idx] = my_list[pivot_idx],my_list[left_bd]
        
        pivot = my_list[left_bd]
        for j in range(left_bd+1,right_bd+1):
            if (my_list[j]<pivot):
                my_list[j],my_list[i] = my_list[i], my_list[j]
                i = i + 1
        my_list[left_bd], my_list[i-1] = my_list[i-1], my_list[left_bd]
        #i is the index of the first element that is bigger than the pivot
        #i-1 is the index of the pivot after the swap
        if i-1 == left_bd:
            right_count = quick_sort_medelem(my_list,left_bd+1,right_bd)
            count+= right_count
        elif i-1 == right_bd:
            left_count = quick_sort_medelem(my_list,left_bd,right_bd-1)
            count+= left_count
        else:
            part_list_left_start = left_bd
            part_list_left_end = i-2
            part_list_right_start = i
            part_list_right_end = right_bd
            left_count = quick_sort_medelem(my_list,part_list_left_start,part_list_left_end)
            right_count = quick_sort_medelem(my_list,part_list_right_start,part_list_right_end)
            count+= (left_count + right_count)
    return count
```
